```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('examscore_dataset.csv')
```

```python
df.head()
```

```python
# a scatter plot comparing IT score and Math score
df.plot(kind='scatter',x='it score',y='math score',color='red')
# plt.show()
```

```python
# Plot column values as a bar plot
# a simple line plot
df.plot(kind='bar',x='country of birth',y='sport score')
```

```python
# Line plot with multiple columns
ax = plt.gca()
df.plot(kind='line',x='name',y='num_children'         ,ax=ax)
df.plot(kind='line',x='name',y='num_pets', color='red', ax=ax)
#
```

```python
# bar plot with group by
# number of unique country 
df.groupby(['country of birth']).size().plot(kind='bar')
```
