```python
# parameter for input image
theImage ='images/girl-2.jpg'
```

```python
# parameter for face detection
theDetectorProto = 'net_models/face_detector/deploy.prototxt'  # model architect
theDetectorModel = 'net_models/face_detector/res10_300x300_ssd_iter_140000.caffemodel' # weight 
theConfidence = 0.7  # minimum probability of detection to be a face # can modify 
theFaceThreshold = 20  # minimum pixels to be a face # can modify
```

```python
# load our serialized face detector from disk 
detector = cv2.dnn.readNetFromCaffe(theDetectorProto, theDetectorModel)
```

```python
# parameter for drawing on the image 
theTextFont = cv2.FONT_HERSHEY_SIMPLEX 
theTextBottomLeft = (10, 100)
theTextScale = 0.85
theTextColor = (255, 0, 255)  # RGB value 
theTextThickness = 2  # text line width
```

```python
# load the image, record image width and height 
#support jpg and png file 
image = cv2.imread(theImage) # read as a numpy array, Blue, Green, Red (BGR) 
(h, w) = image.shape[:2] #show the image width and height 
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # convert BGR to RGB 
plt.imshow(img_rgb) # plot image
```

```python
# construct a blob from the image
imageBlob = cv2.dnn.blobFromImage(cv2.resize(image, (300, 300)), 1.0, (300, 300) #resize the image size
        ,(104.0, 177.0, 123.0), swapRB=False, crop=False) (#change to tensor)
```

```python
# apply OpenCV's deep learning-based face detector to localize faces in the input image
detector.setInput(imageBlob) 
detections = detector.forward() #go through all layers and return the result --> show the possibility
```

```python
# loop over the detections
for i in range(0, detections.shape[2]): 
    #extract the confidence (i.e., probability) associated with the prediction
    prob = detections[0, 0, i, 2]
    # filter out weak detections
    if prob > theConfidence: (#refer to the para : confidence)
        # compute the (x, y) coordinates of the bounding box for the face
        box = detections[0, 0, i, 3:7] * np.array([w, h, w, h]) #find out the X Y 
        (startX, startY, endX, endY) = box.astype("int") #Floaring point in model 
   # extract the face ROI
        face = image[startY:endY, startX:endX]
        (fH, fW) = face.shape[:2]

        # ensure the face width and height are sufficiently large 
        if fW < theFaceThreshold or fH < theFaceThreshold:
            continue

        text = 'Face'
        # draw the bounding box of the face along with the associated probability
        text = "{}: {:.2f}%".format(text, prob * 100)
        y = startY - 10 if startY - 10 > 10 else startY + 10
        cv2.rectangle(image, (startX, startY), (endX, endY), (0, 0, 255), 2) #open cv function 
        cv2.putText(image,text, (startX, y), theTextFont, theTextScale, theTextColor, theTextThickness) 
```

```python
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # convert BGR to RGB
plt.figure (figsize=(16,16) ) #make the plot bigger
plt.imshow(img_rgb) # plot image
```
