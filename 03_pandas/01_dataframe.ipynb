```python
# sort dataframe
lst1 = ['Apple','Orange','Kiwi','Pear','Banana']
lst2 = [18,46,12,9,34]
dict = {"Fruit":lst1, 'Quantity':lst2}
df = pd.DataFrame(dict)
df
```

```python
df.sort_index(ascending=False)
```
