```python
# iloc[row, column] : locate data by position (integer location)
print(df.iloc[0,0], df.iloc[0,1] , df.iloc[0,2])
```

```python
# slice one column
df.iloc[:,1:2]
```

```python
# slice two column
df.iloc[:,1:3]
```

```python
# slice one row
df.iloc[0:1,:]
```

```python
# slice a rectangle
df.iloc[1:3,1:4]
```

```python
# iloc gets rows (or columns) at particular positions in the index (so it only takes integers).
# loc gets rows (or columns) with particular labels from the index.
```

```python
df.loc[2]  ## 2 interpret as label '2', not number 2
```

```python
print(type(df.loc[2]))  # return a Series
```

```python
df.loc[[2,3] ]
```

```python
print(type(df.loc[[2,3]]))  # return a Dataframe
```

```python
# extract selected columns
df.loc[[2,3] , ['gender','country of birth','math score']]
```

```python
# filter USA students by a boolean expression
df.loc[ df['country of birth'] == 'USA' ,['gender','country of birth', 'math score']]
```
