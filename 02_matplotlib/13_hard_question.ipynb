```python
# 先產出數值介於0~5的array
# 並以0.2為間隔，畫出y=x, y=x², y=x³ 的圖表
# r 代表紅色、兩個-代表虛線，
# b代表藍色、s代表方塊
# g代表綠色，^代表三角形
# 詳細指令可按下shift+tab來觀看 docstring
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x = np.arange(0.0 , 5.0, 0.2) 
plt.plot(x, x, 'r--')
```

```python
plt.plot(x, x**2, 'bs')
```

```python
plt.plot(x, x**3, 'g^')
```

```python
plt.plot(x, x, 'r--')
plt.plot(x, x**2, 'bs')
plt.plot(x, x**3, 'g^')
plt.title("combine") 
plt.show() 
```
