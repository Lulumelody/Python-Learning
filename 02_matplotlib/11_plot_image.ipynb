```python
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg
img=mpimg.imread('tiger.png') # support PNG only. With pillow installed. Support loading jpg too.
# return a numpy array with float32, range from 0 to 1
print(type(img))  # class 'numpy.ndarray'
print(img.shape)  # (766,640, 4)  height, width, channel
imgplot = plt.imshow(img)
plt.title('Tiger.png')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg
img=mpimg.imread('tiger.jpg') # support PNG only. return a numpy array with float32, range from 0 to 1
print(type(img))  # class 'numpy.ndarray'
print(img.shape)  #  height, width, channel
imgplot = plt.imshow(img)
plt.title('Tiger.jpg')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2
img = cv2.imread('tiger.jpg')
print(type(img))  # class 'numpy.ndarray'
print(img.shape)  # (150,120,3)  height, width, channel
imgplot = plt.imshow(img)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2
img = cv2.imread('tiger.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print(type(img))  # class 'numpy.ndarray'
print(img.shape)  # (150,120,3)  height, width, channel
imgplot = plt.imshow(img)
plt.axis("off")
plt.show()
```
