```python
import numpy as np 
```

```python
import matplotlib.pyplot as plt 
```

```python
# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 4*np.pi, 0.1) 
y = np.cos(x) 
plt.plot(x, y)
plt.title("cosine wave form") 
plt.xlabel('Radian')
plt.ylabel('num')
plt.show() 
```
