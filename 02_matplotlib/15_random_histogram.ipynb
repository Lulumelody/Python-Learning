```python
import numpy as np 
```

```python
import matplotlib.pyplot as plt 
```

```python
from scipy.stats import norm
```

```python
# plot a histogram
mu = 80  #  mean (average)
sigma = 7  # standard deviation (spread)
x = np.random.normal(mu, sigma, size=200)
```

```python
plt.title('Histogram of 200 random numbers')
plt.hist(x, bins=20, density=True, histtype='bar', color='b', alpha=0.5)
plt.show()
```
