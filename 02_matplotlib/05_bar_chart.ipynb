```python
import matplotlib.pyplot as plt 
```

```python
x = ['a','b','c','d','e']
y = [1,2,3,2,1]
```

```python
plt.bar (x,y)
plt.title("Bar chart") 
plt.show() 
```

```python
plt.bar(x,y)
plt.title("Bar chart")
plt.xlabel('Alphabet')
plt.show()
```

```python
plt.bar(x,y)
plt.title("Bar chart")
plt.xlabel('Alphabet')
plt.ylabel('Value')
plt.show()
```
