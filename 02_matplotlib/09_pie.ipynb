```python
import matplotlib.pyplot as plt 
```

```python
hours  = [8, 8]
activities = ['Sleep', 'Work']
 
plt.pie(hours, labels=activities, colors=['r','g'], startangle=90, autopct='%.1f%%')
plt.show()
```

```python
hours  = [8, 8]
activities = ['Sleep', 'Work']
 
plt.pie(hours, labels=activities, colors=['r','g'], startangle=0, autopct='%.1f%%')
plt.show()
```

```python
hours  = [8, 8, 8]
activities = ['Sleep', 'Work', 'Play']
 
plt.pie(hours, labels=activities, startangle=0, autopct='%.1f%%')
plt.title('My daily activities')
plt.show()
```
