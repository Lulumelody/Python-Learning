** 布林 Booleans **

布林資料類型在 Python 名為`bool`，有`True` 或 `False` 兩個數值。事實上，`bool`是`int`的子類，而`True` 和 `False`兩個名字是`bool`的實例。`True` 和 `False`和數字0, 1是一樣的。

`If`條件式會選擇性地運行代碼，它代表了程式的邏輯。它在 `if` 的格式中是條件，隨後有一個或多個非強制加入的 `elif`條件 (即`否則如果`)，和最後的 `else` 方塊。所有條件都會包含相關的嵌套語句。

當條件式運行時，Python 會評估第一個條件是否真確。如是，有關第一個方塊的程式便會被執行。若所有條件被評估為假，`else`便會被執行。

Boolean data type in Python is called `bool` with two values `True` or `False`. In fact, the names `True` and `False` are instances of `bool`, which is a subclass of the type `int`. `True` and `False` behave the same as the integer 0, 1.

`If` statement select actions to perform and represent the logic of a programme. It takes the form of an `if` test, followed by one or more optional `elif` (meaning "else if") tests and a final option `else` block. All tests are associated with a block of nested statements.



When the if statement runs, Python evaluates if the first test is true. If so, the associated block of the first test is executed. If all the tests are evaluated to be false, the `else` block is executed

### 基本例子 Basic examples

```python
if(4 > 3):
    print("4 is larger than 3")
```

```python
if(4 < 3):
    print("4 is smaller than 3")
else:
    print("4 is larger than 3")
```

** 根據某學生在測試所得的分數計算等級 **

** Calculate the grade for a student based on the score obtained in a test **

```python
score = 60

if(score >= 90):
    grade = 'A'
    
elif(score >= 80):
    grade = 'B'
    
elif(score >= 70):
    grade = 'C'

elif(score >= 50):
    grade = 'D'

else:
    grade = 'F'
    
print('With score %d, the grade obtained is %s' %(score, grade))
```
