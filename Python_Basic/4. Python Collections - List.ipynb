# Python 資料收藏容器
# Python Collections

在 Python 中有4種容器類型可用

1. `字典` - 無序, 可換及有索引的 (沒有重複數值)
2. `列表` - 有序及可換的 (容許重複數值)
3. `集合` - 無序及沒有索引 (沒有重複數值)
4. `元組` - 有序及不可換的 (容許重複數值)

There are four collection data types in Python

1. `Dictionary` - unordered, changeable and indexed (no duplicate members)
2. `List` - ordered and changable (allows duplicate members)
3. `Set` - unordered and unindexed (no duplicate members)
4. `Tuple` - ordered and unchangable (allows duplicate members)


## 列表 Lists

列表是有序及可換的，並於方括號內輸入。在列表中的項目`不一定`是同一種類

List is ordered and changeable and is written with square brackets. Items within the same list `do not` have to be of the same type

```python
example = [1, 2, 3, 4, 'Hello', 'World']
print(example)
```

**提取列表中的數值 Accessing values in Lists **

```python
list1 = ['Amy', 'Betty', 'Christy']
list2 = [18, 22, 24]

print(list1[0])
print(list2[0:2])
      
```

**改變列表中的數值 Changing values in Lists**



```python
list1 = ['Amy', 'Betty', 'Christy']

list1[1] = 'Debbie'

print(list1)
```

**查找某項目是否在列表中 Check if an item exists in the List **


```python
list1 = ['Amy', 'Betty', 'Christy']

if 'Amy' in list1:
    print("Amy exists in the list")
else:
    print("Amy does not exist in the list")
```

**列印在列表中的項目 Print item within the List**

```python
list1 = ['Amy', 'Betty', 'Christy']

for x in list1:
   print(x) 
```

**查看列表長度 Check the length of the list**

```python
list1 = ['Amy', 'Betty', 'Christy']
print(len(list1))
```

**在列表中加插項目 Add item to the list**

有兩種把項目加插至列表的方法。

1. `append()` - 在列表最尾部加入項目
2. `insert()` - 在特定索引中加插項目

There are two ways of adding an item to a list.

1. `append()` - add an item to the end of a list
2. `insert()` - add an item at the specified index

```python
list1 = ['Amy', 'Betty', 'Christy']
list1.append('Debbie')
print(list1)
```

```python
list1 = ['Amy', 'Betty', 'Christy']
list1.insert(1, 'Debbie')
print(list1)
```

**在列表中移除項目 Remove item from the list**

有兩種方法能把列表中的項目移除

1. `remove()` - 把特定數值的項目移除
2. `pop()` - 把特定索引的項目移除

There are two ways of removing an item from the list

1. `remove()` - remove an item with the specified value
2. `pop()` - remove an item at the specified index


```python
list1 = ['Amy', 'Betty', 'Christy']
list1.remove('Betty')
print(list1)
```

```python
list1 = ['Amy', 'Betty', 'Christy']
list1.pop(1)
print(list1)
```

若沒有特別指定索引，系統會默認移除列表最後一項 (即索引-1)

If no parameter is passed, the default is to remove the last item of the list (i.e. index of -1)

```python
list1 = ['Amy', 'Betty', 'Christy']
list1.pop()
print(list1)
```

**把列表項目從小至大排好 Sort the list accendingly**

```python
list1 = [5, 4, 8, 346, 32, 67, 23, 70]
list1.sort()
print(list1)
```

**把列表排序反轉 Reverse the order of a list**

```python
list1 = ['Amy', 'Betty', 'Christy']
list1.reverse()
print(list1)
```

**查找某數值在列表中重複次數 Find the number of elements of a specific value**

```python
list1 = [50, 32 , 34, 56, 83, 3, 83, 23, 20, 94, 83, 97, 78, 72, 72, 92, 71, 61, 39, 40, 73]
print('No of "83" in the list:', list1.count(83))
```

**查找某數值在列表中第一次出現的索引 Find the index of the first element of a specific value**

```python
list1 = [50, 32 , 34, 56, 83, 3, 83, 23, 20, 94, 83, 97, 78, 72, 72, 92, 71, 61, 39, 40, 73]
print('The first "83" in the list:', list1.index(83))
```
