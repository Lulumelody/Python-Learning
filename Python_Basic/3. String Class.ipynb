# 字串 String

`字串` 是 `序列` 的一種，用於記錄文字。序列儲存的物件由左至右掛列。物件會根據它們的掛位儲存和被取出。

`Strings` are used to record textual information and are an example of `sequence`. Sequences maintain a left-to-right order among the items they contain. Items are stored and fetched by their relative position.

```python
s = 'Hello World!'

print(len(s)) # print the length of the string 列印字串長度

print(s[0]) # the first item in s 列印 s 的第一個字元

print(s[1]) # the second item to the left 列印從左開始第二個字元
```

```python
print(s[1:]) # everything past the first (1:len(S)) 第一個字元後所有字

print(s[0:10]) # everything but the last 除最後一個字元外所有字

print(s[:10]) # everything but the last 除最後一個字元外所有字

print(s[:]) # everything 所有字元
```

## 字元方法的例子 String Method Examples 


**字元代替 Replacing text**

我們可用函式`.replace()`

We can use the function `.replace()`

```python
s = 'hello'
s = s[:2] + 'yy' + s[4:]
print(s)

y = 'yellow'
y = y.replace('ll', 'tt')
print(y)
```

**解析文字 Parsing Text**

可用函式`.split()`

`.split()` function can be used 

```python
line = 'xxx yyy zzz'
col1 = line[0:3]
col2 = line[4:7]
col3 = line[8:]

print(col1, col2, col3)

cols = line.split()
print(cols)
```

**大小楷轉換 Case conversions**


```python
s = 'Hello World!'
print(s.lower())
print(s.upper())
```
