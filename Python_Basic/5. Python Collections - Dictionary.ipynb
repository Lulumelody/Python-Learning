## 字典 Dictionary

`字典`是無序, 可換及有索引的容器。它`不會`容許重複項目

A `dictionary` is an unordered, changeable and indexed collection. It `does not` allow duplicate members

```python
person = {'name':'Justin',
          'gender':'Male',
          'age':18,
          'height':178,
          'weight':80,
         }

print(person)
```

**以名字提取項目 Accessing an item by referencing its key name**


```python
# First method 方法一
n = person['name']
print(n)

# Second method 方法二
n = person.get('name')
print(n)
```

**改變某項目的數值 Change the value of a specific item**

```python
person = {'name':'Justin',
          'gender':'Male',
          'age':18,
          'height':178,
          'weight':80,
         }

person['age'] = 20

print(person.get('age'))
```

**列印字典項目 Print item within the dictionary**

```python
#Print all key names 列印所有名字
for x in person:
    print(x)
```

```python
#Print all values - first method 列印所有數值 - 方法一
for x in person:
    print(person[x])
    
#second method 方法二
for x in person.values():
    print(x)
```

**把所有項目和數值循環 Loop through both keys and values**

```python
for x, y in person.items():
    print(x, y)
```

**查看某項目是否存在 Check if a key exists**

```python
person = {'name':'Justin',
          'gender':'Male',
          'age':18,
          'height':178,
          'weight':80,
         }

if('age' in person):
    print('age exists in the dictionary')
else:
    print('age does not exist in the dictionary')
```

**查看字典長度 Check the length of the dictionary**


```python
print(len(person))
```

**在字典中增加項目 Add items to the dictionary**



```python
person = {'name':'Justin',
          'gender':'Male',
          'age':18,
          'height':178,
          'weight':80,
         }

person['education'] = 'Secondary'

print(person)
```

**在字典中移除項目 Remove item from the dictionary**

`pop()` 和 `del` 容許用家以名字把指定項目移除

`pop()` and `del` allow users to remove item by specifying key name

```python
person = {'name':'Justin',
          'gender':'Male',
          'age':18,
          'height':178,
          'weight':80,
         }

person.pop('gender')

print(person)
```

```python
person = {'name':'Justin',
          'gender':'Male',
          'age':18,
          'height':178,
          'weight':80,
         }

del person['gender']

print(person)
```

***注: 若無指定參數，`del`會把整個字典刪除***

***Remarks: `del` can delete the whole dictionary if no parameter is given***

```python
person = {'name':'Justin',
          'gender':'Male',
          'age':18,
          'height':178,
          'weight':80,
         }

del person

print(person)
```

**把字典所有項目刪除 Clear all the items within the dictionary**

```python
person = {'name':'Justin',
          'gender':'Male',
          'age':18,
          'height':178,
          'weight':80,
         }

person.clear()

print(person)
```

**複製字典 Make a copy of the dictionary**

```python
person = {'name':'Justin',
          'gender':'Male',
          'age':18,
          'height':178,
          'weight':80,
         }

person1 = person.copy()

print(person1)
```

**把兩個字典合二為一 Combine two dictionaries together**

```python
person = {'name':'Justin',
          'gender':'Male',
          'age':18,
          'height':178,
          'weight':80,
         }

new_info = {'school':'abc school',
         'score':'100'
        }

person.update(new_info)

print(person)
```
