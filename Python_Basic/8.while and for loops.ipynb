## While迴圈 While Loops

一個 `While` 迴圈中, 只要條件一直是真，便會不斷運行當中的敘述。直至條件是假前，它會一直循環，返回至敘述的開頭。在條件成為假後，敘述會接著`While`方塊。若`While`的條件一開始便是假，當中的敘述永遠不會被運行。

上文的流程圖說明了 `While` 迴圈如何運作。首先，一個變數 `i` 會被指定數值為 1。然後，迴圈會判斷 `i` 是否符合少於或等於10的條件。如果符合，在 `While` 迴圈中的代碼會被執行 - 即列印一個 `'Hello World!'`的字串。然後， `i` 加 1。當條件是假時，程序完結，亦即 `i` 不再少於或等於10。

你可能對每次代碼執行後 `i` 加 1感到疑惑。背後的原因是要避免 `無限迴圈`。`無限迴圈` 指一個 `While` 迴圈的條件永遠不會成為假。 `i` 加 1令程式在重複 11 次後完結。

A `while` loop executes a block of statements as long as a test evaluating to a true value. It keeps looping back to the start of the statement until the test becomes false. The statement follows the `while` block after the test becomes false. If the test of a `while` is false from the beginning, the statement within the `while` block never runs.

The figure above is a logic flow chart illustrating how a `while` loop works. First of all, a variable `i` is assigned with a value of 1. Then, the loop test evaluates if `i` is smaller or equal to 10. If yes, the code within the `while` loop is executed, that is to print a string `'Hello World!'`. Then, `i` is incremented by 1. The process ends when the test becomes false, that is when `i` is not longer smaller or equal to 10.

You may wonder why is `i` incremented by 1 every time when the block of codes is executed. The reason behind is to avoid `infinite loop`. An `infinite loop` refers to the test of a `while` loop never becomes `false`. By incrementing `i` by 1 enables to process to ends after repeating for 11 times.



```python
i = 0

while i <= 10:
    print('Hello World!')
    i = i + 1
```

其他 `While` 迴圈的例子:

Other examples of `while` loop:

```python
x = 0
y = 20

while x < y:
    print(x, end=' ')
    x += 1
    
```

```python
total = 0
money = 0
count = 0

while total < 10000:
    name = input('Please input your name: ')
    money = abs(int(input('Please input the amount for donation: ')))
    total += money
    count += 1
    print('Thank you %s for your donation of $%d' %(name, money))
    
print('Thank you all for supporting! The donation goal is met after %d people donation' %count)
    
```

```python
y = 6
x = y // 2
while x > 1:
    if y % x == 0:
        print(y, 'has factor', x)
        break
    x -= 1
else:
    print(y, 'is prime')
```

## 強制結束, 強制跳出, 忽略 和 Loop else

## break, continue, pass and the Loop else

**強制結束** - 強制跳出整個迴圈

**強制跳出** - 強制跳出本次迴圈，繼續進入下一圈 

**忽略** - 不做任何事情

**Loop else block** - 只有迴圈正常結束時才執行


**break** - Jumps out of the closest enclosing loop (past the entire loop statement)

**continue** - Jumps to the top of the closest enclosing loop (to the loop's header line)

**pass** - Does nothing

**Loop else block** - Runs if and only if it loop is exited normally



執行至`break`敘述時，迴圈會強制結束。

以下例子運用`break`

A loop is exited immediately after the `break` statement is reached.

Example of the use of `break` is provided in the cell below

```python
while True:
    firstname = input('Enter First Name:')
    if firstname == 'stop': break
    lastname = input('Enter Last Name:')
    print('Hello ', firstname, lastname)

print('The while loop is exited after inputting "stop" as First Name')
```

`continue` 敘述會跳至迴圈的頂部。下列例子列印所有少於10和大於或等於10的雙數。

`continue` statement jumps to the top of a loop. The following example prints out all even numbers less than 10 and greater than or equal to 0

```python
a = 10

while a >= 0:
    a -= 1
    if a%2 != 0: continue
    print(a)
```

`Loop else`只有迴圈正常結束時才執行, 即沒有執行 `break` 敘述。下列代碼以計算正整數 `b` 的因素是否大於 1 來判斷它是否質數。

`Loop else` runs if and only if the loop is exited normally, that is without hitting a `break` statement. The following cell determines whether a positive integer `b` is prime by searching for its factor that is greater than 1

```python
b = 7
a = b // 2

while a > 1:
    if b % a == 0:
        print(b, 'has factor', a)
        break
    a -= 1

else:
    print(b, 'is prime')
```

## For 迴圈 For Loops

一個 `for` 迴圈能經過所有排序的物件的項目，它在`字串`，`清單`，`元組`等都能用上。

A `for` loop can step through the items in any ordered sequence object and it works on `strings`, `list`, `tuple`, etc.

```python
x = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']

for day in 3:
    print(day, end = ' ')
```

```python
x = [0, 1, 2, 3, 4, 5]

for num in x:
    print(num)
    
```

### range( ) 函式 The range( ) Function

在`range( )`函式中，一段代碼能循環特定次數

With the `range()` function, a set of code can be looped through by a specificed number of times

```python
for x in range(10):
    print(x)
```

`range( )`函式默認的開始數為 0。加上參數就能改變開始數，如下列例子

The default of `range()` function is 0 as the starting value. The starting value can be changed adding a parameter as shown in the cell below

```python
for x in range(5, 10):
    print(x)
```

`range()`函式默認每次程序加 1。加上第三個參數就能改變開始數，如下列例子

The default of `range()` function is to increment the sequence by 1. The starting value can be changed adding the third parameter as shown in the cell below

```python
for x in range(4, 40, 2):
    print(x)
```

**例子: 計算一個整數的階乘**

**Example: Finding the factorial of an integer**

```python
result = 1

for x in range(1, 11):
    result *= x

print('result = %d' %result)
```

### 巢狀迴圈 Nested Loop

巢狀迴圈是一個迴圈中的迴圈。每次 "外圈" 循環後 "內圈" 就被執行一次

下列例子示範用巢狀迴圈列印一個三角形

A nested loop is a loop inside a loop. The "inner loop" is executed one time for each iteration of the "outer loop" 

The cell below demonstrates the use of nested loops and print a triangle

```python
n = 10
for x in range(1, n + 1):
    for j in range(1, x + 1):
        print('*', end = '')
    print()
```

我們亦能用`巢狀迴圈`列印乘數表

Using `nested loop`, we can also print a multiplication table

```python
for x in range(1, 10):
    for y in range(1, 10):
        print('{0}*{1}={2: ^2}'.format(y, x, x * y), end = " ")
    print()
```
