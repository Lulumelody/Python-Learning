# 模組 Modules

簡單來說，模組便是有 Python 代碼的 `.py` 檔案。模組能設定在其他 Python 檔案也能用上的類別，函式和變數等。

關鍵詞 `import` 和 `from` 可用來使用其他模組

`import`可用來把整個模組搬來
`from`可用來把模組中特定名字搬來

Modules, in simple words, are `.py` files containing Python Code. Modules can be define classes, functions, and variables that can be reference in other Python files.

In order to access other module, keywords `import` and `from` can be used

`import` allows user to fetch a module as a whole
`from` allows user to fetch particular names from a module

```python
import math

#Find the Greatest Common Divisor 尋找最大公因數
x = math.gcd(36, 128)
print(x)
```

```python
#import two modules in the same line 在同一行搬進兩個模組
import math, random

x = math.floor(random.uniform(1,100) * 3.14)
print(x)
```

```python
from math import floor
from random import uniform

x = floor(uniform(1,100) * 3.14)
print(x)
```

## 重新命名模組 Aliasing modules

模組和函式的名字能用關鍵字 `as` 改變

The name of modules and functions can be modified using keyword `as`



```python
import math as m

x = m.pi
print(x)
```

## 運用自定模組 Using a self-defined module

```python
在用模組前，你要先創建一個有 `.py`尾部的 Python 檔案。完成後，你便能把模組搬進程式並開如運用函式。

你可在 `new_module.py` 中更改代碼以加入新函式。

Before you can use a module, you have to first create a Python file that ends with `.py`. When that is done, you can import that module to this programme and access the functions.

You can amend the code in the `new_module.py` to add new functions.
```

```python
import new_module as m

card = m.shuffle()

for x in card:
    print(x)
```
