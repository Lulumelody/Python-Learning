## 元組 Tuple

`元組`是有序而不可換的容器。一個元組被創造後不能增加或移除項目。

A `tuple` is an ordered and unchangable collection. Adding or removing items are not allowed after a tuple is created

**提取元組中的項目 Accessing items in the tuple**

```python
thistuple = ('Amy', 'Betty', 'Charlie')
print(thistuple[0])
```

```python
thistuple = ('Amy', 'Betty', 'Charlie')

for x in thistuple:
    print(x)
```

## 集合 Set 

`集合`是無序且沒有索引的容器。

A `set` is an unordered and unindexed collection.

```python
set1 = {'Amy', 'Betty', 'Charlie'}

print(set1)
```

**提取集合的項目 Access items within a set**

```python
set1 = {'Amy', 'Betty', 'Charlie'}

for x in set1:
    print(x)
```

```python
set1 = {'Amy', 'Betty', 'Charlie'}

#check if 'Amy' exists in the set 查找`Amy`是否在集合中 
print('Amy' in set1)
```

**在集合中增加項目 Add an item to a set**

```python
set1 = {'Amy', 'Betty', 'Charlie'}

set1.add('Debbie')

print(set1)
```

**一次過增加多種項目 Adding multiple items at a time**

```python
set1 = {'Amy', 'Betty', 'Charlie'}

set1.update(['Debbie', 'Eric', 'Fanny'])

print(set1)
```

**從集合移除項目 Remove an item from a set**


```python
set1 = {'Amy', 'Betty', 'Charlie'}

set1.remove('Charlie')

print(set1)
```

```python
set1 = {'Amy', 'Betty', 'Charlie'}

set1.discard('Charlie')

print(set1)
```

**把一個集合空出 Empty a set**

```python
set1 = {'Amy', 'Betty', 'Charlie'}

set1.clear()

print(set1)
```

**移除一個集合 Delete a set**

```python
set1 = {'Amy', 'Betty', 'Charlie'}

del set1

print(set1)
```
