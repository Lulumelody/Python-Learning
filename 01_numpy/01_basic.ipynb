```python
import numpy as np
x = np.array([1, 2, 3]) # Let numpy choose the datatype
print(x)
print(x.dtype)
```

```python
import numpy as np

z = np.array([[1,2,3],[5,6,7]])

print(z)
print(z.dtype)

```

```python
import numpy as np
x = np.array([1.0, 2.0, 3.0])
print(x)
print(x.dtype)
```

```python
import numpy as np
x = np.arange(30)
print(x)
print(x.dtype)
```

```python
import numpy as np
x = np.linspace(0, 4, 4)  # 建立一個 array, 在 0 與 3 的範圍之間 等分 4 個點
print(x)
print(x.dtype)
```

```python
import numpy as np
x = np.array([1, 2])   # Let numpy choose the datatype
print(x.dtype)
```

```python
import numpy as np
x = np.array([1.0, 2.0])   # Let numpy choose the datatype
print(x.dtype) 
```

```python
import numpy as np
x = np.array([1, 2], dtype=np.int64)   # Force a particular datatype
print(x.dtype)
```
