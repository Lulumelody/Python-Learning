```python
import numpy as np
```

```python
a = np.array([1, 2, 3])   # Create a rank 1 array        
```

```python
print(type(a))    # Prints "<class 'numpy.ndarray'>"
```

```python
print(a.shape)            # Prints "(3,)"
```

```python
print(a[0])
print(a[1])
print(a[2])
```

```python
a[0] = 5                  # Change an element of the array
print(a) 
```
