```python
img = cv2.imread('cat-198.jpg')  # Blue, Green, Red
#img = cv2.imread('cat-2.jpg')

# Blue, Green, Red
img_2 = img * [1,    1, 0.2]  # reduce Red. That is more blue/green 
img_3 = img * [1,  0.2, 1]    # reduce green.
img_4 = img * [0.2,  1, 1]    # reduce blue. That is more red/geen

# A slight gotcha with imshow is that it might give strange results
# if presented with data that is not uint8. To work around this, we
# explicitly cast the image to uint8 before displaying it.
#plt.imshow(np.uint8(img_tinted))

img_2 = np.uint8(img_2)
img_3 = np.uint8(img_3)
img_4 = np.uint8(img_4)

# matplotlib use RGB to plot the image
# so, we need to change the color order from BGR to RGB
img   = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_2 = cv2.cvtColor(img_2, cv2.COLOR_BGR2RGB)
img_3 = cv2.cvtColor(img_3, cv2.COLOR_BGR2RGB)
img_4 = cv2.cvtColor(img_4, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(16,16))  # make the plot bigger

plt.subplot(2, 2, 1).set_title('original')   # 2 row2, 2 columns, plot at 1
plt.imshow(img)

plt.subplot(2, 2, 2).set_title('blue/green') # 2 row2, 2 columns plot at 2
plt.imshow(img_2)

plt.subplot(2, 2, 3).set_title('red/blue') # 2 row2, 2 columns plot at 2
plt.imshow(img_3)

plt.subplot(2, 2, 4).set_title('red/green') # 2 row2, 2 columns plot at 2
plt.imshow(img_4)


plt.show()
```
