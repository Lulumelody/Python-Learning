```python
import numpy as np
```

```python
a = np.zeros((3,2))   # Create an array of all zeros
```

```python
print(a)              
```

```python
b = np.ones((3,2))    # Create an array of all ones
print(b)            
```

```python
c = np.full((3,2), 7)  # Create a constant array
print(c)             
```

```python
d = np.eye(3)         # Create a 3x3 identity matrix
print(d)
```

```python
e = np.random.random((3,2))  # Create an array filled with random values
print(e)
```
